# Logical Functios & Lookup



Ques1.Write a formula to check if the marks in cell B2 are greater than or equal to 40. If yes, return "Pass", otherwise "Fail"

    -> Suppose cell range- B2:C6
    Formula: =IF(C2>=40,"Pass","Fail")

Ques2.A company gives a bonus  if sales are above 40,000 and attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not .

    -> Suppose the cell range- B11:D14
    Formula: =IF(And(C11>40000,D11>=90),"Bonus","No Bonus")

Ques3.rite a nested IF formula:
If marks in B2 are ≥ 90 →"A", ≥ 75 → "B", ≥ 50 →"C" , otherwise "Fail".

    -> Suppose cell range- B19:D23
    Formula: =If(C19>=90,"A",If(C19>=75,"B",If(C18>=50,"C","Fail")))
    Answer : B

Ques4.Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?

    -> Suppose cell range- B27:D30
    Formula: =VLOOKUP("E103", B27:D30, 3, False)

    Explanation:
    We use VLOOKUP because the lookup value/EmpId is in the first column of the table. And VLOOKUP is ideal for vertical lookup where data is arranged in columns.
